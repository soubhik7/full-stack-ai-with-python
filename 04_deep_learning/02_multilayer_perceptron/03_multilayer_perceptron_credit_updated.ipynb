
# 📌 Feature Extraction Using `iloc`

```python
inputs = dataset.iloc[:, 1:4].values
```

## 🔎 What This Line Does

This line extracts **input features (X)** from the dataset for training the Multi-Layer Perceptron.

---

## 🧠 Breaking It Down

### 1️⃣ `dataset`
A Pandas DataFrame containing:
- ID columns (if any)
- Feature columns
- Target/output column

---

### 2️⃣ `iloc`
`iloc` means **integer-location based indexing**.

Syntax:

```
dataset.iloc[row_selection, column_selection]
```

It selects rows and columns using **index positions (numbers)** — not column names.

---

### 3️⃣ `:` (Row Selection)

```
:
```

Means:  
✔ Select **ALL rows** in the dataset

---

### 4️⃣ `1:4` (Column Selection)

Python slicing rule:

```
start : stop
```

- Start index → Included  
- Stop index → Excluded  

So `1:4` selects:

- Column index 1  
- Column index 2  
- Column index 3  

It does **NOT** include column 4.

---

### 5️⃣ `.values`

Converts the selected DataFrame into a **NumPy array**.

Why?  
Neural Networks perform matrix operations like:

```
Z = XW
```

Matrix multiplication requires NumPy arrays.

---

## 📐 Resulting Shape

If your dataset has 1000 rows:

```
inputs.shape → (1000, 3)
```

Where:
- 1000 → number of samples
- 3 → number of selected features

---

## 🎯 Why This Is Important

✔ Separates input features from target  
✔ Removes unnecessary columns (like ID)  
✔ Ensures compatibility with matrix multiplication  
✔ Prevents dimension mismatch errors  
✔ Standard preprocessing step in ML pipelines  

---

## 🚀 In Neural Network Context

For example:

```
inputs  → (samples, 3)
weights → (3, hidden_neurons)
```

This allows proper forward propagation.

---

✨ This step is fundamental in preparing structured tabular data for training a Multi-Layer Perceptron.

## Complete neural network

````python
(3 + 1) / 2
````

````python
import numpy as np

def sigmoid(sum):
  return 1 / (1 + np.exp(-sum))

def sigmoid_derivative(sigmoid):
  return sigmoid * (1 - sigmoid)

weights0 = 2 * np.random.random((3, 10)) - 1
weights1 = 2 * np.random.random((10,1)) - 1
````

````python
epochs = 10000
learning_rate = 0.01

error = []

for epoch in range(epochs):
  input_layer = inputs
  sum_synapse0 = np.dot(input_layer, weights0)
  hidden_layer = sigmoid(sum_synapse0)

  sum_synapse1 = np.dot(hidden_layer, weights1)
  output_layer = sigmoid(sum_synapse1)

  error_output_layer = outputs - output_layer
  average = np.mean(abs(error_output_layer))

  if epoch % 1000 == 0:
    print('Epoch: ' + str(epoch + 1) + ' Error: ' + str(average))
    error.append(average)

  derivative_output = sigmoid_derivative(output_layer)
  delta_output = error_output_layer * derivative_output

  weights1T = weights1.T
  delta_output_weight = delta_output.dot(weights1T)
  delta_hidden_layer = delta_output_weight * sigmoid_derivative(hidden_layer)

  hidden_layerT = hidden_layer.T
  input_x_delta1 = hidden_layerT.dot(delta_output)
  weights1 = weights1 + (input_x_delta1 * learning_rate)

  input_layerT = input_layer.T
  input_x_delta0 = input_layerT.dot(delta_hidden_layer)
  weights0 = weights0 + (input_x_delta0 * learning_rate)
````

````python
import matplotlib.pyplot as plt
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.plot(error)
````

````python
def calculate_output(instance):
  hidden_layer = sigmoid(np.dot(instance, weights0))
  output_layer = sigmoid(np.dot(hidden_layer, weights1))
  return output_layer[0]
````
