## Loading the dataset

```python
import tensorflow
tensorflow.__version__
```

```python
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.datasets import fashion_mnist
```

```python
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

```python
i = np.random.randint(0, len(X_train))
plt.imshow(X_train[i], cmap = 'gray')
plt.title('Class ' + str(y_train[i]))
```

```python
X_train.shape
```

```python
X_train = X_train.reshape(60000, 28 * 28)
X_train.shape
```

```python
X_train[0]
```

```python
X_test.shape
```

```python
X_test = X_test.reshape(10000, 28 * 28)
X_test.shape
```

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
X_train /= 255
X_test /= 255
```

```python
X_train.max()
```

```python
X_train.min()
```

```python
y_train
```

## Neural network (training)

```python
(784 + 10) / 2
```

```python
network = Sequential()
network.add(Dense(input_shape = (784,), units = 397, activation = 'relu'))
network.add(Dense(units = 397, activation = 'relu'))
network.add(Dense(units = 10, activation = 'softmax'))
```

```python
network.compile(loss = 'sparse_categorical_crossentropy',
                      optimizer = 'adam',
                      metrics = ['accuracy'])
```

```python
history = network.fit(X_train, y_train, batch_size = 128, epochs = 10)
```

## Neural network (evaluation)

```python
history.history.keys()
```

```python
plt.plot(history.history['loss'])
```

```python
plt.plot(history.history['accuracy'])
```

```python
accuracy_test = network.evaluate(X_test, y_test)
```

```python
predictions = network.predict(X_test)
```
