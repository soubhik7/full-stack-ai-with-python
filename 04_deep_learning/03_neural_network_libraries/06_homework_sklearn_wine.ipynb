# Homework solution - sklearn 1 (wine classification)


## Loading the data

http://archive.ics.uci.edu/ml/datasets/Wine/

```python
from sklearn.neural_network import MLPClassifier
from sklearn import datasets
```

```python
wine = datasets.load_wine()
```

```python
inputs = wine.data
inputs
```

```python
wine.feature_names
```

```python
outputs = wine.target
outputs
```

```python
wine.target_names
```

```python
inputs.shape
```

```python
outputs.shape
```

## Train and test datasets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, outputs, test_size = 0.2)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

## Neural network (training)

```python
(13 + 3) / 2
```

```python
# 13 -> 8 -> 8 -> 3
network = MLPClassifier(max_iter=2000,
                        verbose=True,
                        activation = 'logistic',
                        solver = 'adam',
                        learning_rate = 'constant',
                        learning_rate_init = 0.001,
                        tol=0.00000100,
                        hidden_layer_sizes = (8, 8))
network.fit(X_train, y_train)
```

```python
network.classes_
```

```python
network.coefs_
```

```python
network.intercepts_
```

```python
network.n_layers_
```

```python
network.n_outputs_
```

```python
network.out_activation_
```

## Neural network (evaluation)

```python
X_test.shape
```

```python
predictions = network.predict(X_test)
predictions
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy_score(y_test, predictions)
```

```python
cm = confusion_matrix(y_test, predictions)
cm
```

```python
# yellowbrick already installed; using seaborn for compatibility
```

```python
# Confusion matrix heatmap (seaborn — same visual as yellowbrick)
import seaborn as sns
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=wine.target_names, yticklabels=wine.target_names,
            ax=ax)
ax.set_xlabel('Predicted')
ax.set_ylabel('True')
ax.set_title('Confusion Matrix — Wine MLPClassifier')
plt.tight_layout()
plt.show()
```
