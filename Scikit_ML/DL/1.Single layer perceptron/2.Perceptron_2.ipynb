# Perceptron 2

## Inputs and weights

```python
import numpy as np
```

```python
inputs = np.array([35, 25])
```

```python
type(inputs)
```

```python
weights = np.array([-0.8, 0.1])
```

## Sum function

```python
def sum(inputs, weights):
  return inputs.dot(weights)
```

```python
s = sum(inputs, weights)
```

```python
s
```

## Step function

```python
def step_function(sum):
  if (sum >= 1):
    return 1
  return 0
```

## Final result

```python
step_function(s)
```
