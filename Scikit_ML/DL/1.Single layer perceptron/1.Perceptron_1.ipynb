## Sum function

```python
def sum(inputs, weights):
  s = 0
  for i in range(2):
    #print(i)
    #print(inputs[i])
    #print(weights[i])
    s += inputs[i] * weights[i]
    
  return s
```

```python
def weighted_sum(inputs, weights):
    print("Step 1: Function started")
    print("Inputs:", inputs)
    print("Weights:", weights)
    
    s = 0
    print("\nStep 2: Initialize sum (s = 0)")
    
    for i in range(2):
        print("\n----------------------------")
        print("Loop iteration:", i)
        
        print("Current input value:", inputs[i])
        print("Current weight value:", weights[i])
        
        multiplication = inputs[i] * weights[i]
        print("Multiplication (input * weight):", multiplication)
        
        s += multiplication
        print("Updated sum (s):", s)
    
    print("\nStep 3: Loop finished")
    print("Final sum:", s)
    
    return s


# Example usage
inputs = [2, 3]
weights = [4, 5]

result = weighted_sum(inputs, weights)
print("\nReturned value:", result)
```
