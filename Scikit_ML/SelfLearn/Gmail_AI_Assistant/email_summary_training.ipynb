# Email Summary Model

This notebook demonstrates how to build an extractive email summarizer using Scikit-Learn tools (TF-IDF) alongside SpaCy for sentence parsing. We define a custom Scikit-Learn transformer that can `fit` and `predict` (or `transform`) raw email texts into summaries.

Once initialized, we save the model using `joblib` so it can be loaded in our GitHub Actions pipeline.

```python
import os
import spacy
import pickle
import joblib

# Import the custom summarizer we created
from extractive_summarizer import ExtractiveSummarizer
```

### Initialize the Model

Our model is unsupervised, meaning we don't need a large tagged dataset to train it. TF-IDF will automatically score sentences based on keyword importance within each single email!


```python
model = ExtractiveSummarizer(top_n_sentences=2)
```

### Test the Summarizer with Dummy Data


```python
sample_email = """Dear team, I am writing to inform you that the quarterly meeting has been pushed back to next week. \
The reason for this delay is that our CEO will be traveling and we want her to be present. \
Please update your calendars and make sure to prepare your respective reports by next Wednesday. \
Also, do not forget to submit your expense reports by tomorrow. \
Thank you for your cooperation!"""

summary = model.predict([sample_email])[0]
print("Original Text:\n", sample_email, "\n")
print("\n--- SUMMARY ---\n")
print(summary)
```

### Save the Model

Now we save the instance to a `model.pkl` file to load into `summarize.py`.

```python
with open('model.pkl', 'wb') as f:
    pickle.dump(model, f)
print("Model saved successfully as model.pkl!")
```
