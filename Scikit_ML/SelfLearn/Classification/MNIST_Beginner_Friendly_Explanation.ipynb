```python

# STEP 1: Import Required Libraries

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml

print("Libraries imported successfully!")
```


## STEP 2: Load the MNIST Dataset

Scikit-Learn datasets can return:

1. Generated datasets â†’ `(X, y)` tuple  
2. Real datasets â†’ `Bunch` object (like a dictionary)

A Bunch object contains:
- `data` â†’ input features
- `target` â†’ labels
- `DESCR` â†’ description

For image datasets like MNIST, we set `as_frame=False`
to get NumPy arrays instead of Pandas DataFrames.


```python

print("Loading MNIST dataset...")

mnist = fetch_openml("mnist_784", version=1, as_frame=False)

print("Dataset loaded successfully!")
```


## STEP 3: Extract Input Data (X) and Labels (y)

- `X` â†’ input features (pixel values)
- `y` â†’ target labels (digits 0â€“9)


```python

X = mnist.data
y = mnist.target

print("X shape:", X.shape)
print("y shape:", y.shape)
```


## Why 784 Features?

Each MNIST image:
- 28 pixels wide
- 28 pixels tall

Total pixels = 28 Ã— 28 = 784

Each pixel has intensity from:
- 0 â†’ white
- 255 â†’ black

So each image becomes a row of 784 numbers.



## STEP 4: Display One Digit

Currently, each image is a flat array of 784 values.

We reshape it back to 28 Ã— 28 to display it.


```python

# Take first image
some_digit = X[0]

# Reshape to 28x28
some_digit_image = some_digit.reshape(28, 28)

print("Label of this digit:", y[0])

plt.imshow(some_digit_image, cmap="binary")
plt.axis("off")
plt.show()
```


## Understanding Shapes

- `X.shape = (70000, 784)`  
  â†’ 70,000 images  
  â†’ 784 features per image  

- `y.shape = (70000,)`  
  â†’ 70,000 labels  

Each row in `X` corresponds to one value in `y`.
