```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
connect_token = ["connecting", "connected", "connection", "connections", "connects"]

for t in connect_token:
    print(f"{t} --> {ps.stem(t)}")
```

```python
learn_token = ["learning", "learned", "learner", "learns", "learned", "learn"]
for t in learn_token:
    print(f"{t} --> {ps.stem(t)}")
```

```python
likes_token = ["liking", "liked", "liker", "likes", "like", "better", "worse"]
for t in likes_token:
    print(f"{t} --> {ps.stem(t)}")
```
