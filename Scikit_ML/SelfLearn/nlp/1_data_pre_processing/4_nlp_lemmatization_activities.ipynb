```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```python
connect_token = ["connecting", "connected", "connection", "connections", "connects", "connectivity"]
learn_token = ["learning", "learned", "learner", "learns", "learned", "learn"]
likes_token = ["liking", "liked", "liker", "likes", "like", "better", "worse"]
```

```python
lemmatizer = WordNetLemmatizer()

for word in connect_token:
    print(f"{word} --> {lemmatizer.lemmatize(word)}")
```

```python
for word in learn_token:
    print(f"{word} --> {lemmatizer.lemmatize(word)}")
```

```python
for word in likes_token:
    print(f"{word} --> {lemmatizer.lemmatize(word)}")
```
