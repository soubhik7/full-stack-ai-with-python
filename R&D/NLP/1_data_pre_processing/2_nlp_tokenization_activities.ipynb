```python
import nltk
nltk.download('punkt_tab') # punkt_tab : a table used to figure out where sentences start and end
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
sentences = "Hi This is Soubhik, and Soubhik cat's name is Luna. His dog's name is Max."
```

```python
sent_tokenize(sentences)
```

```python
sentence = " her cat's name is Luna"
word_tokenize(sentence)
```

```python
sentence_2 = "Her cat's name is Luna and her dog's name is Max."
word_tokenize(sentence_2)
```
