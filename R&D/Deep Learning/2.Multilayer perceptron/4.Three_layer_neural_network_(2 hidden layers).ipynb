```python
import numpy as np

# -----------------------------
# Activation Function
# -----------------------------

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(output):
    # derivative assumes sigmoid output already computed
    return output * (1 - output)

# -----------------------------
# Example Dataset (Replace with yours)
# -----------------------------

inputs = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

outputs = np.array([[0], [1], [1], [0]])

outputs = outputs.reshape(-1, 1)

# -----------------------------
# Network Architecture
# -----------------------------

n_input = inputs.shape[1]
n_hidden1 = 8
n_hidden2 = 6
n_output = 1

# Xavier Initialization for stable gradients
weights0 = np.random.randn(n_input, n_hidden1) * np.sqrt(1 / n_input)
weights1 = np.random.randn(n_hidden1, n_hidden2) * np.sqrt(1 / n_hidden1)
weights2 = np.random.randn(n_hidden2, n_output) * np.sqrt(1 / n_hidden2)

# -----------------------------
# Hyperparameters
# -----------------------------

epochs = 100000
learning_rate = 0.6

# -----------------------------
# Training Loop
# -----------------------------

for epoch in range(1, epochs + 1):

    # =====================
    # Forward Propagation
    # =====================

    input_layer = inputs

    # Layer 1
    z1 = np.dot(input_layer, weights0)
    hidden1 = sigmoid(z1)

    # Layer 2
    z2 = np.dot(hidden1, weights1)
    hidden2 = sigmoid(z2)

    # Output Layer
    z3 = np.dot(hidden2, weights2)
    output_layer = sigmoid(z3)

    # =====================
    # Loss (Mean Absolute Error)
    # =====================

    error_output = outputs - output_layer
    loss = np.mean(np.abs(error_output))

    # =====================
    # Backpropagation
    # =====================

    # ---- Output Layer Gradient ----
    delta_output = error_output * sigmoid_derivative(output_layer)

    # ---- Hidden Layer 2 Gradient ----
    error_hidden2 = delta_output.dot(weights2.T)
    delta_hidden2 = error_hidden2 * sigmoid_derivative(hidden2)

    # ---- Hidden Layer 1 Gradient ----
    error_hidden1 = delta_hidden2.dot(weights1.T)
    delta_hidden1 = error_hidden1 * sigmoid_derivative(hidden1)

    # =====================
    # Weight Updates (Gradient Descent)
    # =====================

    # Update weights2
    weights2 += hidden2.T.dot(delta_output) * learning_rate

    # Update weights1
    weights1 += hidden1.T.dot(delta_hidden2) * learning_rate

    # Update weights0
    weights0 += input_layer.T.dot(delta_hidden1) * learning_rate

    # =====================
    # Accuracy Calculation
    # =====================

    if epoch % 10000 == 0:

        predictions = (output_layer >= 0.5).astype(int)
        correct = np.sum(predictions == outputs)
        total = outputs.shape[0]
        accuracy = correct / total

        print(
            f"Epoch: {epoch} "
            f"Loss: {loss:.15f} "
            f"Accuracy: {accuracy:.15f} "
            f"({correct}/{total})"
        )

# -----------------------------
# Final Results
# -----------------------------

print("\nFinal Probabilities:")
print(output_layer)

print("\nFinal Predictions:")
print((output_layer >= 0.5).astype(int))
```

```python
import numpy as np

# -----------------------------
# Activation Functions
# -----------------------------

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return (x > 0).astype(float)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(output):
    return output * (1 - output)

# -----------------------------
# Example Dataset (XOR)
# -----------------------------

inputs = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

outputs = np.array([[0], [1], [1], [0]])
outputs = outputs.reshape(-1, 1)

# -----------------------------
# Network Architecture
# -----------------------------

n_input = inputs.shape[1]
n_hidden1 = 8
n_hidden2 = 6
n_output = 1

# He Initialization (important for ReLU)
weights0 = np.random.randn(n_input, n_hidden1) * np.sqrt(2 / n_input)
weights1 = np.random.randn(n_hidden1, n_hidden2) * np.sqrt(2 / n_hidden1)
weights2 = np.random.randn(n_hidden2, n_output) * np.sqrt(2 / n_hidden2)

# -----------------------------
# Hyperparameters
# -----------------------------

epochs = 100000
learning_rate = 0.01   # lower LR recommended for ReLU stability

# -----------------------------
# Training Loop
# -----------------------------

for epoch in range(1, epochs + 1):

    # =====================
    # Forward Propagation
    # =====================

    # Layer 1
    z1 = np.dot(inputs, weights0)
    hidden1 = relu(z1)

    # Layer 2
    z2 = np.dot(hidden1, weights1)
    hidden2 = relu(z2)

    # Output Layer
    z3 = np.dot(hidden2, weights2)
    output_layer = sigmoid(z3)

    # =====================
    # Loss (Mean Absolute Error)
    # =====================

    error_output = output_layer - outputs
    loss = np.mean(np.abs(error_output))

    # =====================
    # Backpropagation
    # =====================

    # Output layer gradient
    delta_output = error_output * sigmoid_derivative(output_layer)

    # Hidden layer 2 gradient
    delta_hidden2 = delta_output.dot(weights2.T) * relu_derivative(z2)

    # Hidden layer 1 gradient
    delta_hidden1 = delta_hidden2.dot(weights1.T) * relu_derivative(z1)

    # =====================
    # Weight Updates (Gradient Descent)
    # =====================

    weights2 -= hidden2.T.dot(delta_output) * learning_rate
    weights1 -= hidden1.T.dot(delta_hidden2) * learning_rate
    weights0 -= inputs.T.dot(delta_hidden1) * learning_rate

    # =====================
    # Accuracy Calculation
    # =====================

    if epoch % 10000 == 0:

        predictions = (output_layer >= 0.5).astype(int)
        correct = np.sum(predictions == outputs)
        total = outputs.shape[0]
        accuracy = correct / total

        print(
            f"Epoch: {epoch} "
            f"Loss: {loss:.15f} "
            f"Accuracy: {accuracy:.15f} "
            f"({correct}/{total})"
        )

# -----------------------------
# Final Results
# -----------------------------

print("\nFinal Probabilities:")
print(output_layer)

print("\nFinal Predictions:")
print((output_layer >= 0.5).astype(int))
```
