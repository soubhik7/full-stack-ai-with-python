# 📘 Deep Learning Notes: Parameters vs Hyperparameters (Detailed)

---

## 1. 🧠 Core Idea

In deep learning, there are **two fundamentally different types of variables**:

### 1. Parameters (learned)
### 2. Hyperparameters (set by you)

👉 Entire training process revolves around:

```
Hyperparameters → control learning → produce Parameters → define model
```

---

## 2. 🔑 Parameters (W, b)

### Definition
Parameters are the **internal variables** of the model that are **learned automatically** during training.

### Examples
- Weights: $$ W^{[l]} $$
- Biases: $$ b^{[l]} $$

### Mathematical Role

Forward propagation:
$$
Z^{[l]} = W^{[l]} A^{[l-1]} + b^{[l]}
$$

Activation:
$$
A^{[l]} = g(Z^{[l]})
$$

👉 These directly determine predictions:
$$
\hat{y} = A^{[L]}
$$

---

### How Parameters are Learned

Using **Gradient Descent**:

$$
W := W - \alpha \frac{\partial J}{\partial W}
$$

$$
b := b - \alpha \frac{\partial J}{\partial b}
$$

👉 Goal: Minimize cost function $$ J(W, b) $$

### Key Properties

- Automatically updated
- Data-dependent
- Final output of training
- Cannot be manually chosen

---

## 3. ⚙️ Hyperparameters

### Definition
Hyperparameters are **external configuration variables** that control the **learning process**.

👉 They are NOT learned — you must set them.

---

### Why “Hyper”?

Because they operate at a **higher level**:
- They control how parameters evolve
- They influence optimization behavior

---

## 4. 📊 Types of Hyperparameters

### A. Optimization Hyperparameters
- Learning rate (\( \alpha \))
$$
\alpha
$$
- Number of iterations / epochs
- Mini-batch size

### B. Model Architecture Hyperparameters
- Number of layers (\( L \))
- Number of hidden units 
$$
n^{[l]}
$$
- Activation functions:
  - ReLU
  - Sigmoid
  - Tanh

### C. Regularization Hyperparameters (Advanced)
- L2 regularization (\( \lambda \))
- Dropout rate
- Early stopping criteria

### D. Advanced Optimization Hyperparameters
- Momentum
- Adam parameters (β1, β2)
- Learning rate decay

---

## 5. 🔁 Relationship Between Them

### Key Insight
Hyperparameters **indirectly determine** the final parameters.

👉 Example:
- Large learning rate → unstable updates → bad \( W \)
- Good learning rate → smooth convergence → optimal \( W \)

---

## 6. 📉 Learning Rate (α) – Critical Hyperparameter

### Behavior of Cost Function \( J \)

- Small α → very slow convergence
- Large α → divergence
- Optimal α → fast and stable convergence

### Intuition
> Learning rate is like step size while descending a hill

- Too small → slow 🐢  
- Too large → unstable 🚫  
- Just right → efficient ✅  

---

## 7. 🔄 Empirical Nature of Deep Learning

> Deep learning is an **empirical process**

- No exact formula for best hyperparameters
- Requires experimentation

---

### Trial-and-Error Loop

```
1. Choose hyperparameters
2. Train model
3. Evaluate performance
4. Adjust hyperparameters
5. Repeat
```

---

## 8. 🧪 Hyperparameter Search

### Common strategies:
- Manual tuning
- Grid search
- Random search

---

## 9. 🌍 Real-World Observations

### Domain Dependency
Hyperparameters vary across:
- Computer Vision
- NLP
- Speech
- Recommendation systems

### Changing Over Time
- New data
- Better hardware
- New architectures

👉 Re-tune periodically

---

## 10. 🧠 Developing Intuition

- Start with guesses
- Improve with experience

---

## 11. 📊 Evaluation Strategy

- Training set → learn parameters
- Validation set → tune hyperparameters

👉 Never tune on test set

---

## 12. ⚠️ Common Challenges

- Large search space
- Expensive computation
- No guaranteed best solution

---

## 13. 🔥 Why So Many Hyperparameters?

- Deep models are complex
- More flexibility → more tuning

---

## 14. 🧾 Summary Table

| Aspect | Parameters | Hyperparameters |
|--------|----------|----------------|
| Definition | Learned | Set manually |
| Examples | W, b | α, L |
| Updated by | Gradient descent | User |
| Role | Prediction | Control learning |

---

## 15. 🎯 Final Takeaways

- Parameters = what model learns  
- Hyperparameters = how model learns  
- Deep learning = experimentation-heavy  
- Success = tuning + intuition  


## ➕ Parameter Update (Layer-wise Form)

$$
W^{[l]} = W^{[l]} - \alpha \text{ } dW^{[l]} \tag{16}
$$

$$
b^{[l]} = b^{[l]} - \alpha \text{ } db^{[l]} \tag{17}
$$

Where:
$$
\alpha = \text{learning rate}
$$

$$
dW^{[l]}, \; db^{[l]} = \text{gradients for layer } l
$$
