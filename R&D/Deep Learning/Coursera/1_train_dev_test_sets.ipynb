## The Iterative Process in Deep Learning

Applied machine learning is highly iterative:
1. Start with an idea (network architecture, hyperparameters, etc.)
2. Implement and run an experiment
3. Evaluate results on dev set
4. Refine ideas and iterate

Key decisions include:
- Number of layers
- Hidden units per layer
- Learning rate
- Activation functions

Intuitions from one domain (e.g., NLP) don't always transfer to others (e.g., computer vision). The best choices depend on:
- Amount of data
- Input features
- Compute resources (GPUs/CPUs)

Setting up data correctly speeds up this iterative cycle.

## Setting Up Train/Dev/Test Sets

### Traditional Approach
- **Train set**: Used to train algorithms
- **Dev set** (development/validation set): Used to evaluate different models and select the best one
- **Test set**: Used for final unbiased evaluation of the selected model

### Traditional Splits
- Small datasets: 70/30 (train/test) or 60/20/20 (train/dev/test)
- These ratios were common when datasets had 100-10,000 examples

### Modern Approach for Large Datasets
- With millions of examples, dev and test sets can be much smaller
- Example: 1 million examples → 98% train, 1% dev, 1% test (10,000 each for dev/test)
- Even smaller: 99.5% train, 0.25% dev, 0.25% test

**Guideline**: Dev set needs to be large enough to evaluate algorithm choices (e.g., 10 different configurations). Test set needs to give confident performance estimates.

## Mismatched Train/Test Distributions

### Common Scenario
- Train on data from one source (e.g., high-quality images from web)
- Dev/Test on data from target application (e.g., user-uploaded blurry photos)

### Key Rule
**Dev and Test sets should come from the same distribution**

Why?
- You optimize heavily on dev set
- Dev set should represent the target performance you care about
- Test set gives unbiased estimate of that performance

### Training Data Strategy
- Use creative methods to get more training data (e.g., web crawling)
- It's okay if train data differs from dev/test, as long as dev/test match
- This allows faster progress despite distribution mismatch

## When It's Okay to Not Have a Test Set

### Purpose of Test Set
- Provides unbiased performance estimate of final model
- Not always necessary

### Alternative Setup
- Train on training set
- Evaluate and iterate on dev set only
- No separate test set

### Terminology Note
- Some teams call dev set "test set" but use it for iterative development
- This leads to overfitting on the "test" set
- Better to use proper terminology: train/dev vs train/dev/test

### When to Skip Test Set
- If you don't need unbiased performance estimate
- Common in industry when rapid iteration is prioritized over final unbiased metrics

## Benefits of Proper Data Setup

- Faster iteration in the development cycle
- Better ability to measure bias and variance
- More efficient hyperparameter tuning and model selection
- Clear separation between development and final evaluation
